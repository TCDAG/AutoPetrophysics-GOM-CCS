```python
#projectile motion with air resistance

import numpy as np
import matplotlib.pyplot as plt

grav = 9.8
airDen = 1.225      # kg/m^3
areaProj = 0.8643   # m^2
v0 = 198.1          # m/s
angle = 60.0   
dragCoef = 0.62     #Approximate value
mass = 15.0         #kg
step = 0.02    #time step in seconds 

t = [0]
vx = [v0*np.cos(angle*np.pi/180)]
vy = [v0*np.sin(angle*np.pi/180)]
x = [0]
y = [0]

dragForce = 0.5*dragCoef*areaProj*(v0**2)*airDen


ax = [-(dragForce*np.cos(angle/180*np.pi))/mass]
ay = [-grav-(dragForce*np.sin(angle/180*np.pi)/mass)]


counter = 0
while(y[counter] >= 0):
    t.append(t[counter]+step)

    vx.append(vx[counter]+step*ax[counter])
    vy.append(vy[counter]+step*ay[counter])

    x.append(x[counter]+step*vx[counter])
    y.append(y[counter]+step*vy[counter])

    vel = np.sqrt(vx[counter+1]**2 + vy[counter+1]**2)
    dragForce = 0.5*dragCoef*areaProj*(vel**2)*airDen

    angle = np.arctan(vy[counter]/vx[counter]) * (180 / np.pi)

    ax.append(-(dragForce*np.cos(angle/180*np.pi))/mass)
    ay.append(-grav-(dragForce*np.sin(angle/180*np.pi)/mass))

    counter=counter+1


plt.plot(x,y)
plt.ylabel("y (m)")
plt.xlabel("x (m)")
```

```python
#projectile motion without air resistance
#https://stackoverflow.com/questions/59258178/fixing-projectile-motion-formulas-in-python

#In the absence of air resistance, the horizontal motion is completely independent of the vertical motion.
#Therefore, a projectile in motion will remain with the same horizontal velocity throughout its flight,
#since no force exists in the horizontal direction, 
#but its velocity will change in the vertical direction due to the force of gravity.

#Read more: Ballistics - Projectile Motion Without Air Resistance - Cannon, Horizontal, Direction, and Force - JRank Articles https://science.jrank.org/pages/727/Ballistics-Projectile-motion-without-air-resistance.html#ixzz7oZZyLfnR

grav = 9.8
airDen = 1.225      # kg/m^3
areaProj = 0.8643   # m^2
v0 = 198.1          # m/s
angle = 60.0   
dragCoef = 0.62     #Approximate value
mass = 15.0         #kg
step = 0.02    #time step in seconds 

t = [0]
vx = [v0*np.cos(angle*np.pi/180)]
vy = [v0*np.sin(angle*np.pi/180)]
x1 = [0]
y1 = [0]

dragForce = 0.5*dragCoef*areaProj*(v0**2)*airDen


ax = [-grav-(np.sin(angle/180*np.pi)/mass)]
ay = [-grav-(dragForce*np.sin(angle/180*np.pi)/mass)]


counter = 0
while(y1[counter] >= 0):
    t.append(t[counter]+step)

    vy.append(vx[counter]+step*ax[counter])
    vy.append(vy[counter]+step*ay[counter])

    x1.append(x1[counter]+step*vx[counter]) 
    y1.append(y1[counter]+step*vy[counter])

    vel = np.sqrt(vx[counter+1]**2 + vy[counter+1]**2)
    dragForce = 0.5*dragCoef*areaProj*(vel**2)*airDen

    angle = np.arctan(vy[counter]/vx[counter]) * (180 / np.pi)

    #ax.append(-(dragForce*np.cos(angle/180*np.pi))/mass)
    ay.append(-grav-(dragForce*np.sin(angle/180*np.pi)/mass))

    counter=counter+1


plt.plot(x1,y1)
plt.ylabel("y1 (m)")
plt.xlabel("x1 (m)")

```

```python
import sys
from tabulate import tabulate
headers = ['time (s)', 'x1', 'y1']    


table = zip(t, x1, y1)
print(tabulate(table, headers=headers, floatfmt=".4f"))
```
