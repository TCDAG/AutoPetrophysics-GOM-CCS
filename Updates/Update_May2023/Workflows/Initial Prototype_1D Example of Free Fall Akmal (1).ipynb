```python
#https://physics.stackexchange.com/questions/104368/free-falling-object-with-air-resistance

import numpy as np
import matplotlib.pyplot as plt

coeff=0.2

times = np.linspace(0,5,100)
tstep = (max(times)-min(times))/len(times)

no_drag_velocities = []
for i in times:
    no_drag_velocities = 9.8*times

#print(no_drag_velocities)


a=0
v=0
drag_velocities = []
for i in times:
    drag_velocities.append(v)
    a=9.8-coeff*v**2*tstep
    v=v+a*tstep

#print(drag_velocities)

plt.plot(times,no_drag_velocities,label="Without Drag") 
plt.plot(times,drag_velocities,label="With Drag")
plt.title('Velocity vs time ')
plt.xlabel("Time [s]")
plt.ylabel("Speed [m/s]")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
import sys
from tabulate import tabulate
headers = ['time (s)', 'velocity without drag force', 'velocity with drag velocity']    

#!{sys.executable} -m pip install tabulate
x = times
y1 = no_drag_velocities
y2= drag_velocities
#print (discrepancy)

table = zip(x, y1, y2)
print(tabulate(table, headers=headers, floatfmt=".4f"))
```

```python
x_min = min(x); x_max = max(x);


norm_x = (x - x_min)/(x_max - x_min) * 2 - 1
print(norm_x)
```

```python
y2 = np.asarray(y2)
#type(y2)
y2_min = min(y2); y2_max = max(y2);
norm_y2 = (y2 - y2_min)/(y2_max - y2_min) * 2 - 1


print(norm_y2)
```

```python
import pandas.util.testing as tm                       # DataFrames for tabular data
import os                                 # set working directory, run executables
import seaborn as sns                     # for plotting
import warnings                           # supress warnings from seaborn pairplot
from sklearn.model_selection import train_test_split # train / test DatFrame split
(x_train, x_test, y_train, y_test) = train_test_split(norm_x, norm_y2, test_size=0.25, random_state=73073)
```

```python
plt.subplot(121)
plt.plot(x_train,y_train, 'o', markerfacecolor='red', markeredgecolor='black', alpha=0.4, label = "Train")
plt.plot(x_test,y_test, 'o', markerfacecolor='blue', markeredgecolor='black', alpha=0.4, label = "Test")
plt.legend(loc='upper right');
plt.title('Velocity vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.xlim(-1,1); plt.ylim(-1,1)

plt.subplots_adjust(left=0.0, bottom=0.0, right=2.2, top=1.2, wspace=0.2, hspace=0.2)
plt.show()


#plt.plot(x1_test,x2_test, 'o')
#plt.xlabel('Without Drag'); plt.ylabel('With Drag')
#plt.grid()
#plt.show()
```

```python
import tensorflow as tf
from keras.models import Sequential, load_model
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils
import keras
from tensorflow.python.keras import backend as k

# Build Neural Network for prediction model
model_1 = Sequential([
    Dense(1, activation='relu', input_shape=(1,)),
    Dense(3, activation='relu'),
    Dense(5, activation='relu'),
    Dense(3, activation='relu'),
    Dense(1, activation='tanh'),
])

tf.keras.optimizers.Adam(learning_rate=0.003, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

model_1.compile(optimizer='adam',
              loss='mse')

tf.keras.optimizers.Adam(learning_rate=0.003, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

model_1.compile(optimizer='adam',
              loss='mse')

#training the network
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)

hist_1 = model_1.fit(x_train, y_train,
          batch_size=5, epochs=300,
          validation_data=(x_test, y_test))
```

```python
# Specify the prediction locations
nbins = 100
x_bins = np.linspace(x_min, x_max, nbins) # set the bins for prediction
norm_x_bins = (x_bins-x_min)/(x_max-x_min)*2-1 # use normalized bins


#predicting using the network
pred_norm_y = model_1.predict(np.array(norm_x_bins)) # predict with our ANN
pred_Ny = ((pred_norm_y + 1)/2*(y2_max - y2_min)+y2_min)

x_train = ((x_train + 1)/2*(x_max - x_min)+x_min)
y_train = ((y_train + 1)/2*(y2_max - y2_min)+y2_min)
x_test = ((x_test + 1)/2*(x_max - x_min)+x_min)
y_test = ((y_test + 1)/2*(y2_max - y2_min)+y2_min)


print (pred_Ny)


# Plot the Model Predictions
plt.subplot(1,1,1)
plt.plot(x_bins,pred_Ny,'black',linewidth=2)
plt.plot(x_train,y_train, 'o', markerfacecolor='red', markeredgecolor='black', alpha=1.0, label = "Train")
plt.plot(x_test, y_test, 'o', markerfacecolor='blue', markeredgecolor='black', alpha=1.0, label = "Test")
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.subplots_adjust(left=0.0, bottom=0.0, right=2.0, top=0.8, wspace=0.2, hspace=0.2)
plt.show()
```

```python
plt.subplot(1,1,1)
plt.plot(hist_1.history['loss'])
plt.plot(hist_1.history['val_loss'])
plt.title('Loss function of Model 1')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.tight_layout()
```
